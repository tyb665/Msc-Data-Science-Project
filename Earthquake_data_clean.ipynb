```python
import pandas as pd
```

Merge the earthquake data and the comcat data.

```python
# Read original source files to generate a clean combined dataset
comcat_df = pd.read_csv(r"D:\project data\MSc_Data_Science_Project\comcat_catalog.csv")
earthquakes_df = pd.read_csv(r"D:\project data\MSc_Data_Science_Project\earthquakes.csv")

# Prepare comcat_cleaned to match earthquakes structure
comcat_cleaned = comcat_df[[
    'longitude', 'latitude', 'depth', 'magnitude', 'time_string'
]].copy()
comcat_cleaned.rename(columns={'time_string': 'time'}, inplace=True)
comcat_cleaned['type'] = 'eq'
comcat_cleaned['magnitude_type'] = 'l'

# Ensure time is properly parsed
comcat_cleaned['time'] = pd.to_datetime(comcat_cleaned['time'], errors='coerce')
earthquakes_df['time'] = pd.to_datetime(earthquakes_df['time'], errors='coerce')

# Combine both datasets
combined_clean_df = pd.concat([earthquakes_df, comcat_cleaned], ignore_index=True)

# Drop rows with missing time values
combined_clean_df.dropna(subset=['time'], inplace=True)

# Save to file
output_path = r"D:\project data\MSc_Data_Science_Project\combined_earthquakes_cleaned.csv"
combined_clean_df.to_csv(output_path, index=False)

combined_clean_df
```
